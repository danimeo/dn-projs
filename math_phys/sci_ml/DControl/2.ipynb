```python
import control
from control import tf, series, parallel, feedback, step_response
import numpy as np
import matplotlib.pyplot as plt

# (2) 
Ke=6.0/0.1047e3
Kt=57.3e-3
Ra=1.00
Jm=0.325e-4
JL=16*Jm
J=JL+Jm

Km=1/Ke;
Tm=Ra*J/(Ke*Kt)
M=tf(Km, [Tm, 1])
t, y = step_response(M)
plt.plot(t, y)
plt.show()

# (2) 
KOmN=1/0.1047
n0=4000
KV=n0/(Km*KOmN)

V=tf(KV, 1)
OmN=tf(KOmN, 1)
sys=V*M*OmN
t, y = step_response(sys)
plt.plot(t, y)
plt.show()

# (3)
KF=2.0/0.1047e3
Uf=1.0

# KA
KA=Uf/(KF*(n0*0.1047))


KP=[1, 5, 10, 50]
Ki=1.0

for Kp in KP:
    P=tf(1,1)
    I=tf(Ki, [1, 0])
    PI=tf(Kp,1)*parallel(P,I)
    sys=feedback(PI*V*M, KF*KA, -1)*OmN
    print(sys)
    t=np.arange(0, 2, 0.01)
    t, y = step_response(sys, t)
    plt.plot(t, y)
plt.show()

# KpKi

Kp=8.9501
Ki=0.0001

sys=feedback(tf([Kp, Kp*Ki],[1, 0])*V*M, KF*KA, -1)*OmN;
print(sys)
t=np.arange(0, 2, 0.01)
t, y = step_response(sys, t)
plt.plot(t, y)
plt.show()

sys=tf([1800, 0.18], [0.1683, 1.45, 4.501e-5])
t, y = step_response(sys, t)
plt.plot(t, y)
plt.show()
sys=tf([1.18e4, 1.18], [0.1683, 3.95, 0.000295])
t, y = step_response(sys, t)
plt.plot(t, y)
plt.show()
sys=tf([3800, 3611], [0.1683, 1.95, 0.9027])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()
sys=tf([5800, 8411], [0.1683, 2.45, 2.103])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()
sys=tf([9800, 2.401e4], [0.1683, 3.45, 6.003])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()
sys=tf([1.38e4, 4.761e4], [0.1683, 4.45, 11.9])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()
sys=tf([3.9e4, 2.301e5], [0.1683, 10.75, 57.53])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()

sys=tf([4.08e4, 2.407e5], [0.1683, 11.2, 60.18])
t, y = step_response(sys)
plt.plot(t, y)
plt.show()

Kp=10.2
Ki=5.9
sys=feedback(tf([Kp, Kp*Ki],[1, 0])*V*M, KF*KA, -1)*OmN;
print(sys)
t=np.arange(0, 2, 0.01)
t, y = step_response(sys, t)
plt.plot(t, y)
plt.show()




```
